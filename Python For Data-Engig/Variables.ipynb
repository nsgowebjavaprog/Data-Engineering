# Variables

### 🔹 1. What is a Variable in Python?

A **variable is a reference (label) to an object in memory**, not the object itself.

```python
x = 10
```

* `10` → actual object in memory
* `x` → just a reference pointing to that object

👉 Python uses **dynamic typing**, so no need to declare type.

---

### 🔹 2. Dynamic Typing (Very Important)

```python

x = "data"    # now string
```

* Type is attached to **object**, not variable
* Useful in ETL pipelines where schema may vary

---

### 🔹 3. Multiple Assignment

```python

```

```python

```

Used in:

* initializing configs

---

### 🔹 4. Memory Management & Identity

```python
a = 10
b = 10
```

```python
id(a) == id(b)  # True (small integer caching)
```

* Python may reuse memory for immutable objects
* Important for **performance in large data processing**

---

### 🔹 5. Mutable vs Immutable (🔥 Most Asked)

### Immutable (cannot change)

```python
x = 10
x = x + 1   # new object created
```

Types:

* int, float, string, tuple

---

### Mutable (can change)

```python
lst = [1, 2, 3]
lst.append(4)   # same object modified
```

Types:

* list, dict, set

---

### ⚠️ Interview Trick

```python
a = [1, 2]
b = a
b.append(3)

print(a)  # [1, 2, 3]
```

👉 Both point to **same object**

### 🔹 6. Shallow vs Deep Copy (Data Engineering Use Case)

### Shallow Copy

```python
import copy
a = [[1, 2], [3, 4]]
b = copy.copy(a)
```

* Outer object copied
* Inner objects shared

---

### Deep Copy

```python
b = copy.deepcopy(a)
```

* Full independent copy

👉 Important when:

* transforming nested JSON
* handling Spark/Pandas transformations

## 🔹 Shallow Copy ()

1. Shallow copy creates a **new outer object** but shares **inner object references**.
2. Nested objects are **not copied**, only their references are used.
3. So, modifying inner data affects **both original and copy**.
4. It is **fast and memory efficient**.
5. Created using `.copy()`, slicing `[:]`, or `copy.copy()`.

### Example:

```python
a = [[1, 2], [3, 4]]
b = a.copy()

b[0][0] = 100
print(a)  # [[100, 2], [3, 4]]  (changed)
```

---

## 🔹 Deep Copy ()

1. Deep copy creates a **fully independent object**, including all nested objects.
2. It recursively duplicates **every level of data**.
3. Changes in copy **do not affect original**.
4. It is **slower and uses more memory**.
5. Created using `copy.deepcopy()`.

### Example:

```python
import copy

a = [[1, 2], [3, 4]]
b = copy.deepcopy(a)

b[0][0] = 100
print(a)  # [[1, 2], [3, 4]]  (unchanged)
```
