## Strings

### 🔹 1. What is a String?

* A string is an **immutable sequence of characters**.
* Defined using `' '`, `" "`, or `''' '''`.

```python
s = "data engineering"
```

👉 Stored as a sequence → supports indexing & slicing.

---

### 🔹 2. Immutability (🔥 Most Important)

* Strings **cannot be changed in-place**.

```python
s = "hello"
s[0] = "H"   # ❌ Error
```

✔ Instead:

```python
s = "H" + s[1:]
```

👉 New string object is created.

---

### 🔹 3. Indexing & Slicing

```python
s = "python"
```

* Indexing:

```python
s[0]   # 'p'
s[-1]  # 'n'
```

* Slicing:

```python
s[0:3]   # 'pyt'
s[::-1]  # 'nohtyp' (reverse)
```

---

### 🔹 4. String Methods (Common + Important)

### Case Conversion

```python
s.upper()
s.lower()
s.title()
```

---

### Cleaning Data (Data Engineering Use Case)

```python
s.strip()      # remove spaces
s.replace("a", "x")
```

---

### Splitting & Joining (🔥 Very Important)

```python
s = "a,b,c"
s.split(",")   # ['a','b','c']
```

```python
",".join(["a","b"])   # 'a,b'
```

👉 Used in:

* CSV processing
* log parsing
* ETL pipelines

---

### 🔹 5. Searching & Checking

```python

s.startswith("da")
s.endswith("ing")
```

---

### Validation Methods

```python
s.isalpha()
s.isdigit()
s.isalnum()
```

👉 Used for data validation pipelines

---

### 🔹 6. String Formatting (🔥 Interview Favorite)

### Old Style

```python
"Hello %s" % name
```

### format()

```python
"Hello {}".format(name)
```

### f-string (Best)

```python
f"Hello {name}"
```

👉 Fastest & recommended

---

### 🔹 7. String Interning & Memory (Advanced)

```python
a = "hello"
b = "hello"

a is b   # True (interned)
```

👉 Python reuses memory for small strings → optimization

---

### 🔹 8. Important Concepts for Interviews

## 🔸 `is` vs `==`

```python
a = "hi"
b = "hi"

a == b   # True
a is b   # True (sometimes)
```

---

#### 🔸 String Concatenation Cost

```python
s = ""
for i in range(10000):
    s += "a"   # ❌ slow
```

✔ Better:

```python
"".join(["a"] * 10000)
```
