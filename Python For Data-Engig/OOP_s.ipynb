```python
class cls_name():
    company = "ABC Pvt Ltd"   # class variable
    
    def __init__(self, name):
        self.name = name
    
    def info_func(self):
        print(f'Employee name is {self.name}')
    
    @classmethod
    def change_company(cls, new_name):
        cls.company = new_name   # modifies class variable

# usage
obj1 = cls_name('John')
obj2 = cls_name('Alice')

print(cls_name.company)   # ABC Pvt Ltd

cls_name.change_company("XYZ Ltd")

print(obj1.company)  # XYZ Ltd
print(obj2.company)  # XYZ Ltd
```

## Getters & Setters

Why use getters & setters?

Add validation before setting values

Hide internal implementation (encapsulation)

Control how data is accessed/modified

```python
class Employee:
    def __init__(self, salary):
        self._salary = salary   # protected variable

    @property
    def salary(self):
        return self._salary     # getter

    @salary.setter
    def salary(self, value):
        if value > 0:           # validation
            self._salary = value
        else:
            print("Invalid salary")

# usage
emp = Employee(50000)

print(emp.salary)   # get value
emp.salary = 60000  # set value
emp.salary = -10    # invalid
```

```python
class Employee:
    def __init__(self, name, salary):
        self._name = name        # protected variable (by convention)
        self._salary = salary    # protected variable
    
    # Getter for name
    def get_name(self):
        return self._name        # return name value
    
    # Setter for name
    def set_name(self, name):
        self._name = name        # update name
    
    # Getter for salary
    def get_salary(self):
        return self._salary      # return salary
    
    # Setter for salary with validation
    def set_salary(self, salary):
        if salary > 0:           # validation condition
            self._salary = salary
        else:
            print("Salary must be positive!")

# Usage
emp = Employee("John", 50000)

print(emp.get_name())     # access name using getter
emp.set_name("Alice")     # update name using setter

print(emp.get_salary())   # access salary
emp.set_salary(60000)     # valid update
emp.set_salary(-100)      # invalid update (handled)

print(emp.get_salary())
```
