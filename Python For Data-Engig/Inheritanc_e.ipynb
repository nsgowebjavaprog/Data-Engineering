## Inheritance

### 🔹 What is Inheritance?

Inheritance allows a **child class (subclass)** to reuse properties and methods of a **parent class (superclass)**.

It helps in:

* Code reuse
* Reducing redundancy
* Building scalable data pipelines

---

## 🔹 Basic Syntax

```python
class Parent:
    pass

class Child(Parent):
    pass
```

---

### 🔹 Simple Example (Interview Level)

```python
# Parent class
class DataPipeline:
    def __init__(self, source):
        self.source = source

    def extract(self):
        print(f"Extracting data from {self.source}")

# Child class
class ETLPipeline(DataPipeline):
    def transform(self):
        print("Transforming data")

    def load(self):
        print("Loading data into warehouse")

# Usage
obj = ETLPipeline("AWS S3")
obj.extract()   # inherited
obj.transform()
obj.load()
```

**Interview Insight:**
Here, `ETLPipeline` reuses extraction logic and adds transformation + loading — similar to real-world ETL pipelines.

### 🔹 Types of Inheritance (Important for Interviews)

#### 1. Single Inheritance

```python
class A:
    pass

class B(A):
    pass
```

---

#### 2. Multiple Inheritance

```python
class A:
    def func1(self):
        print("A")

class B:
    def func2(self):
        print("B")

class C(A, B):
    pass
```

👉 Used rarely in data engineering due to complexity.

---

#### 3. Multilevel Inheritance

```python
class A:
    pass

class B(A):
    pass

class C(B):
    pass
```

---

#### 4. Hierarchical Inheritance

```python
class A:
    pass

class B(A):
    pass

class C(A):
    pass
```

---

### 🔹 Method Overriding (Very Important)

Child class can **modify parent behavior**.

```python
class DataPipeline:
    def process(self):
        print("Basic processing")

class CustomPipeline(DataPipeline):
    def process(self):
        print("Custom processing for big data")

obj = CustomPipeline()
obj.process()
```

**Interview Tip:**
Used when different pipelines need customized transformations.

---

### 🔹 super() Function (Must Know)

Used to call parent class constructor/method.

```python
class DataPipeline:
    def __init__(self, source):
        self.source = source

class ETLPipeline(DataPipeline):
    def __init__(self, source, target):
        super().__init__(source)
        self.target = target
```

### 🔹 Real Data Engineering Use Case

```python
class BaseIngestion:
    def read_data(self):
        print("Reading raw data")

class KafkaIngestion(BaseIngestion):
    def read_data(self):
        print("Reading data from Kafka")

class S3Ingestion(BaseIngestion):
    def read_data(self):
        print("Reading data from S3")
```

This is how frameworks handle **multiple data sources** cleanly.

### 🔹 Key Interview Points (Must Remember)

* Inheritance = **IS-A relationship**

  * Example: ETLPipeline **is a** DataPipeline

* Promotes:

  * Code reuse
  * Modularity
  * Scalability

* Method Resolution Order (MRO):

```python
print(Child.__mro__)
```

* Avoid deep inheritance → use composition when needed

### 🔹 Common Interview Questions

**1. Difference between inheritance & composition?**

* Inheritance → IS-A
* Composition → HAS-A

**2. Why avoid multiple inheritance?**

* Complexity, ambiguity (Diamond Problem)

**3. What is method overriding?**

* Redefining parent method in child class

**4. What is super()?**

* Calls parent class methods

> Inheritance in Python allows a child class to reuse properties and methods of a parent class. It supports code reusability and modular design, which is useful in building scalable data pipelines. Python supports multiple types of inheritance like single, multiple, and multilevel. Method overriding allows customization of inherited behavior. The `super()` function is used to access parent class methods.

## Adding constructors-->  Inheritance

````python
class ns_company:
    
    def __init__(self, com_name):
        self.com_name = com_name

    def display_info(self):
        print(f"Company Name: {self.com_name}")
        
class employee(ns_company):    
        
    def __init__(self, com_name, emp_name):
        super().__init__(com_name)   # ✅ call parent constructor - super(key)
        self.emp_name = emp_name

    def display_emp_info(self):
        print(f"Employee Name: {self.emp_name}")
        
emp_1 = employee("NS Corp", "Alice Smith")

emp_1.display_info()       # from parent
emp_1.display_emp_info()  # from child
````

### 
🔹 1. Single Inheritance

````python
# One parent → One child

class Parent:
    def show(self):
        print("This is Parent class")

class Child(Parent):  # Child inherits Parent
    def display(self):
        print("This is Child class")

obj = Child()
obj.show()      # inherited method
obj.display()
````

### 
🔹 2. Multiple Inheritance

````python
# One child → Multiple parents

class Father:
    def skill1(self):
        print("Father: Driving")

class Mother:
    def skill2(self):
        print("Mother: Cooking")

class Child(Father, Mother):  # inherits both
    def skill3(self):
        print("Child: Coding")

obj = Child()
obj.skill1()   # from Father
obj.skill2()   # from Mother
obj.skill3()
````

### 🔹 3. Multilevel Inheritance

````python
# Grandparent → Parent → Child

class Grandparent:
    def gp(self):
        print("Grandparent class")

class Parent(Grandparent):
    def p(self):
        print("Parent class")

class Child(Parent):
    def c(self):
        print("Child class")

obj = Child()
obj.gp()  # from Grandparent
obj.p()   # from Parent
obj.c()
````

🔹 4. Hierarchical Inheritance

````python
# One parent → Multiple children

class Parent:
    def common(self):
        print("Common method from Parent")

class Child1(Parent):
    def c1(self):
        print("Child1 method")

class Child2(Parent):
    def c2(self):
        print("Child2 method")

obj1 = Child1()
obj2 = Child2()

obj1.common()
obj1.c1()

obj2.common()
obj2.c2()
````

🔹 5. Hybrid Inheritance (Combination)

````python
# Combination of multiple + hierarchical

class A:
    def a(self):
        print("Class A")

class B(A):
    def b(self):
        print("Class B")

class C(A):
    def c(self):
        print("Class C")

class D(B, C):  # Hybrid (multiple + multilevel)
    def d(self):
        print("Class D")

obj = D()
obj.a()  # from A
obj.b()  # from B
obj.c()  # from C
obj.d()
````

🔥 Bonus: Method Overriding (Very Important)

````python
class Parent:
    def show(self):
        print("Parent method")

class Child(Parent):
    def show(self):  # overriding parent method
        print("Child method")

obj = Child()
obj.show()  # Child method (overrides parent)
````

🔥 Bonus: Constructor Inheritance

````python
class Parent:
    def __init__(self):
        print("Parent constructor")

class Child(Parent):
    def __init__(self):
        Parent.__init__(self)  # calling parent constructor
        print("Child constructor")

obj = Child()
````

Single → 1 parent, 1 child
Multiple → multiple parents
Multilevel → chain (A → B → C)
Hierarchical → one parent, many children
Hybrid → combination
Overriding → child changes parent method
Constructor → call parent using super() or direct call

````python
# ================================
# ALL TYPES OF INHERITANCE IN ONE FILE
# ================================

# 🔹 Base Class (used for many examples)
class A:
    def method_A(self):
        print("Method from Class A")


# ================================
# 🔹 1. Single Inheritance
# One parent → One child
# ================================
class B(A):  # B inherits A
    def method_B(self):
        print("Method from Class B")

print("\n--- Single Inheritance ---")
obj1 = B()
obj1.method_A()  # inherited from A
obj1.method_B()


# ================================
# 🔹 2. Multiple Inheritance
# One child → Multiple parents
# ================================
class C:
    def method_C(self):
        print("Method from Class C")

class D(A, C):  # D inherits from A and C
    def method_D(self):
        print("Method from Class D")

print("\n--- Multiple Inheritance ---")
obj2 = D()
obj2.method_A()  # from A
obj2.method_C()  # from C
obj2.method_D()


# ================================
# 🔹 3. Multilevel Inheritance
# Grandparent → Parent → Child
# ================================
class E(A):  # Parent
    def method_E(self):
        print("Method from Class E")

class F(E):  # Child
    def method_F(self):
        print("Method from Class F")

print("\n--- Multilevel Inheritance ---")
obj3 = F()
obj3.method_A()  # from A (grandparent)
obj3.method_E()  # from E (parent)
obj3.method_F()


# ================================
# 🔹 4. Hierarchical Inheritance
# One parent → Multiple children
# ================================
class G(A):  # Child 1
    def method_G(self):
        print("Method from Class G")

class H(A):  # Child 2
    def method_H(self):
        print("Method from Class H")

print("\n--- Hierarchical Inheritance ---")
obj4 = G()
obj5 = H()

obj4.method_A()
obj4.method_G()

obj5.method_A()
obj5.method_H()


# ================================
# 🔹 5. Hybrid Inheritance
# Combination (multiple + multilevel)
# ================================
class I(B, C):  # inherits B (which already inherits A) + C
    def method_I(self):
        print("Method from Class I")

print("\n--- Hybrid Inheritance ---")
obj6 = I()
obj6.method_A()  # from A (via B)
obj6.method_B()  # from B
obj6.method_C()  # from C
obj6.method_I()


# ================================
# 🔥 Bonus: Method Overriding
# ================================
class Parent:
    def show(self):
        print("Parent show method")

class Child(Parent):
    def show(self):  # overriding
        print("Child show method")

print("\n--- Method Overriding ---")
obj7 = Child()
obj7.show()  # calls child version


# ================================
# 🔥 Bonus: Constructor Inheritance
# ================================
class Parent2:
    def __init__(self):
        print("Parent2 constructor")

class Child2(Parent2):
    def __init__(self):
        Parent2.__init__(self)  # calling parent constructor
        print("Child2 constructor")

print("\n--- Constructor Inheritance ---")
obj8 = Child2()
````
