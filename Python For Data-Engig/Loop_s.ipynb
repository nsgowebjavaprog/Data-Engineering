---

### 🔹 What are Loops?

Loops are used to **execute a block of code repeatedly** over a sequence or until a condition is met.

👉 In Data Engineering, loops are used for:

* Iterating over datasets
* Processing records
* Automating ETL steps
* Batch processing

---

### 🔹 Types of Loops in Python

1. `for` loop
2. `while` loop

---

### 🔹 1. `for` Loop

Used to iterate over **iterables** like lists, tuples, strings, dictionaries.

### Syntax

```python
for item in iterable:
    # code block
```

### Example (Data Engineering context)

```python
data = [10, 20, 30]

for value in data:
    print(value * 2)
```

👉 Used in:

* Reading records from datasets
* Iterating over files, APIs, logs

---

### 🔹 Looping with `range()`

```python
for i in range(5):
    print(i)
```

👉 Generates sequence → `0 to n-1`
👉 Used in batch processing or indexing

---

### 🔹 Looping through Dictionary

```python
record = {"name": "John", "age": 25}

for key, value in record.items():
    print(key, value)
```

👉 Common in JSON data processing

---

### 🔹 2. `while` Loop

Executes as long as **condition is True**.

### Syntax

```python
while condition:
    # code block
```

### Example

```python
count = 0

while count < 3:
    print(count)
    count += 1
```

👉 Used in:

* Streaming data
* Waiting for conditions (e.g., file arrival)

---

### 🔹 Loop Control Statements

### 1. `break`

Stops loop immediately

```python
for i in range(5):
    if i == 3:
        break
```

---

### 2. `continue`

Skips current iteration

```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

---

### 3. `pass`

Does nothing (placeholder)

```python
for i in range(5):
    pass
```

---

### 🔹 Nested Loops

Loop inside another loop

```python
for i in range(2):
    for j in range(2):
        print(i, j)
```

👉 Used in:

* Multi-dimensional data
* Matrix / table processing

---

### 🔹 Real Data Engineering Use Case

```python
data = [{"age": 25}, {"age": None}, {"age": -5}]

for record in data:
    if record["age"] is None:
        print("Missing")
    elif record["age"] < 0:
        print("Invalid")
    else:
        print("Valid")
```

👉 Common in:

* Data validation
* Data cleaning pipelines

---

### 🔹 Performance Note (Important for Interviews ⚡)

* Python loops are **slow for large data**
* Prefer:

  * **Vectorized operations (Pandas)**
  * **Distributed processing (PySpark)**

👉 Interview Tip:
**“Avoid loops when working with big data; use optimized frameworks like Spark.”**

---

### 🔹 One-Line Strong Interview Answer

**“Loops in Python (for, while) are used for iterative execution over data, and in data engineering they are mainly applied in record processing, ETL workflows, and automation, though for large-scale data we prefer distributed tools like PySpark instead of raw loops.”**
