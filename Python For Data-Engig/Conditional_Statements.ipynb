## 🔹 What are Conditional Statements?

Conditional statements are used to **control the flow of execution** based on conditions (True/False).

In Data Engineering, they are heavily used in:

* Data validation
* Data filtering
* ETL pipelines (decision making)
* Handling missing or incorrect data

---

## 🔹 Syntax Overview

```python
if condition:
    # block executes if condition is True
elif another_condition:
    # executes if above is False and this is True
else:
    # executes if all conditions are False
```

---

## 🔹 1. `if` Statement

Executes code **only when the condition is True**.

### Example (Data Engineering context)

```python
records = 100

if records > 0:
    print("Data available for processing")
```

👉 Used when checking:

* File exists or not
* Data is empty or not
* Threshold validations

---

## 🔹 2. `elif` Statement

Used to check **multiple conditions sequentially**.

### Example

```python
records = -5

if records > 0:
    print("Valid data")
elif records == 0:
    print("No data")
else:
    print("Invalid data")
```

👉 Important:

* Only the **first True condition executes**
* Rest are skipped

---

## 🔹 3. `else` Statement

Executes when **all above conditions are False**.

### Example

```python
status = "failed"

if status == "success":
    print("Pipeline completed")
else:
    print("Pipeline failed")
```

👉 Used for:

* Default handling
* Error cases

---

## 🔹 4. Nested Conditions

Condition inside another condition.

```python
age = 25
country = "India"

if age > 18:
    if country == "India":
        print("Eligible user")
```

👉 Used in:

* Complex data rules
* Multi-level validations

---

## 🔹 5. Conditional Expressions (One-line)

```python
result = "Valid" if records > 0 else "Invalid"
```

👉 Useful in:

* Data transformations
* Quick assignments in pipelines

---

## 🔹 6. Logical Operators in Conditions

```python
if records > 0 and records < 1000:
    print("Valid batch size")
```

Operators:

* `and` → both True
* `or` → at least one True
* `not` → reverse condition

---

## 🔹 Real Data Engineering Use Case

```python
data = {"name": "John", "age": None}

if data["age"] is None:
    print("Missing value detected")
elif data["age"] < 0:
    print("Invalid age")
else:
    print("Valid record")
```

👉 This is typical in:

* Data cleaning
* Schema validation
* ETL pipelines

---

## 🔹 Interview-Level Key Points

* Python uses **indentation (not braces)**
* Conditions evaluate to **Boolean (True/False)**
* `elif` avoids multiple nested `if` blocks
* Short-circuiting happens in logical operators
* Widely used in **Airflow DAGs, PySpark transformations, and validation logic**
