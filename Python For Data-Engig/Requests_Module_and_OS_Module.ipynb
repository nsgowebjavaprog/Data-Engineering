### 🔹 1. `requests` Module (HTTP/API Handling)

#### 👉 What is it?

`requests` is used to **send HTTP requests** (GET, POST, PUT, DELETE) to APIs.

👉 In Data Engineering:

* Fetch data from APIs (REST APIs)
* Ingest external data (weather, finance, etc.)

---

## ##🔹 Basic Example (GET Request)

```python
import requests

response = requests.get("https://jsonplaceholder.typicode.com/posts")

print(response.status_code)   # HTTP status
print(response.json()[0])     # First record
```

---

#### 🔹 POST Request Example

```python
import requests

data = {
    "title": "Data Engineer",
    "body": "Python + SQL",
    "userId": 1
}

response = requests.post("https://jsonplaceholder.typicode.com/posts", json=data)

print(response.status_code)
print(response.json())
```

---

#### 🔹 Important Concepts

* `response.status_code` → 200 = success
* `response.json()` → converts API data to Python dict
* `headers` → authentication (API keys)

---

#### 🔹 Real Data Engineering Example

```python
import requests

url = "https://api.github.com/users"

res = requests.get(url)

if res.status_code == 200:
    data = res.json()
    print("Total users fetched:", len(data))
```

---

#### 🧠 Interview Points

* Used for **API ingestion pipelines**
* Handles **REST APIs easily**
* Supports authentication, headers, params

# 🔹 2. `os` Module (Operating System Interaction)

#### 👉 What is it?

`os` module is used to **interact with the file system and OS**

👉 In Data Engineering:

* File handling
* Directory management
* Pipeline automation

---

#### 🔹 Basic Example

```python
import os

print(os.getcwd())  # current working directory
```

---

#### 🔹 File & Directory Operations

```python
import os

# create folder
os.mkdir("data_folder")

# list files
print(os.listdir())

# rename folder
os.rename("data_folder", "new_data_folder")

# delete folder
os.rmdir("new_data_folder")
```

---

#### 🔹 Path Handling (Important)

```python
import os

file_path = os.path.join("data", "file.csv")
print(file_path)
```

---

#### 🔹 Environment Variables (Very Important)

```python
import os

# get environment variable
print(os.getenv("PATH"))

# set environment variable
os.environ["MY_KEY"] = "12345"
```

👉 Used for:

* API keys
* Database credentials

---

#### 🔹 Real Data Engineering Example

```python
import os

folder = "data_files"

if not os.path.exists(folder):
    os.mkdir(folder)

print("Files:", os.listdir(folder))
```

---

### 🔥 Difference: `requests` vs `os`

| Feature | requests   | os             |
| ------- | ---------- | -------------- |
| Purpose | API calls  | OS interaction |
| Usage   | Fetch data | Manage files   |
| Domain  | Networking | System         |
| Example | GET/POST   | mkdir, listdir |

---

### 🔹 Combined Example (Real Pipeline)

```python
import requests
import os

# create folder
if not os.path.exists("api_data"):
    os.mkdir("api_data")

# fetch data
res = requests.get("https://jsonplaceholder.typicode.com/posts")

# save to file
with open("api_data/data.txt", "w") as f:
    f.write(str(res.json()))

print("Data saved successfully")
```

---

### 🧠 5-Line Interview Answer

> The `requests` module is used to handle HTTP requests and is commonly used in data engineering to fetch data from APIs. The `os` module is used to interact with the operating system, such as file handling, directory management, and environment variables. Requests is mainly used for external data ingestion, while os is used for managing local system resources. Both are essential in building end-to-end data pipelines.
